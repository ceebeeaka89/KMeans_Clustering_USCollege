```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop('Private',axis=1))
```

```python
kmeans.cluster_centers_
```

```python
import matplotlib.pyplot as plt

# Assuming 'kmeans_labels' are the cluster labels assigned by KMeans
plt.scatter(df['Enroll'], df['Outstate'], c=kmeans.labels_, cmap='viridis', edgecolor='k', s=50)
plt.title('Scatter Plot of KMeans Labels')
plt.show()
```
